```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("scholarship.csv")
```

```python
df1=df.copy()
```

```python
df1.head()
```

```python
df1.shape
```

```python
sns.displot(df1["semester_percentage"],kde=True)
sns.displot(df1["scholarship_exam_marks"],kde=True)
plt.show()
```

```python
mean=df1["semester_percentage"].mean()
std=df1["semester_percentage"].std()
```

```python
print(f"mean:{mean}\nstandard deviation:{std}")
```

```python
highest_allowed=mean+3*std
lowest_allowed=mean-3*std
```

```python
print(f"Highest allowed percentage:{highest_allowed}\nLowest allowed percentage:{lowest_allowed}")
```

```python
df1[(df1["semester_percentage"]<lowest_allowed)|(df1["semester_percentage"]>highest_allowed)]
```

## Trimming

```python
df1[(df1["semester_percentage"]>lowest_allowed)&(df1["semester_percentage"]<highest_allowed)]
```

## using z-score

```python
df1["z_score"]=(df1["semester_percentage"]-df1["semester_percentage"].mean())/df1["semester_percentage"].std()
```

```python
df1
```

```python
df1[(df1["z_score"]<-3) | (df1["z_score"]>3)]
```

```python
df1[(df1["z_score"]>-3) & (df1["z_score"]<3)]
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("weight-height.csv")
```

```python
df
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df["Height"])
plt.subplot(2,2,2)
sns.boxplot(df["Height"])
plt.show()
```

```python
min_limit=df["Height"].quantile(0.01)
max_limit=df["Height"].quantile(0.99)
print(f"minimum limit:{min_limit}\nmaximum limit:{max_limit}")
```

```python
df[(df["Height"]>max_limit)|(df["Height"]<min_limit)]
```

## Trimming

```python
tdf=df[(df["Height"]<max_limit)&(df["Height"]>min_limit)]
```

```python
tdf
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(tdf["Height"])
plt.subplot(2,2,2)
sns.boxplot(tdf["Height"])
plt.show()
```

## Capping

```python
df["Height"]=np.where(df["Height"]>max_limit,max_limit,np.where(df["Height"]<min_limit,min_limit,df["Height"]))
```

```python
df
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df["Height"])
plt.subplot(2,2,2)
sns.boxplot(df["Height"])
plt.show()
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head()
```

## 1. Removing the columns with missing data

```python
df1=df.copy()
df1.head()
```

```python
df1.isna().sum()
```

```python
df1.drop("last_review",axis=1,inplace=True)
```

```python
df1.isna().sum()
```

## 2. Deleting rows with missing values

```python
df2=df.copy()
```

```python
df2.isna().sum()
```

```python
df2.dropna(inplace=True)
```

```python
df2.isna().sum()
```

```python
df2.shape
```

## 3. Filling Missing values - Imputation(Numerical data)

```python
df3=df.copy()
```

```python
mean_value=df3["reviews_per_month"].mean()
df3["reviews_per_month"] = df3["reviews_per_month"].fillna(mean_value)
```

```python
df3.head()
```

```python
df3.isna().sum()
```

## 3. Filling missing values - Imputation(Categorical data)

```python
df3.nunique()
```

```python
df3["last_review"]=df3["last_review"].fillna(df3["last_review"].value_counts().index[0])
```

```python
df3.isna().sum()
```

```python
df4=df.copy()
```

```python
df4.isna().sum()
```

```python
df4["last_review"]=df4["last_review"].fillna("Not Reviewed")
```

```python
df4.isna().sum()
```

```python
df4.head()
```

## 4. Advanced Imputation

```python
df5=df.copy()
```

```python
df5.isna().sum()
```

```python
df5["reviews_per_month"]=df5["reviews_per_month"].interpolate()
```
